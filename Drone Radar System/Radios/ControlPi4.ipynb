```python
!pip install pymavlink
!pip install pyserial
from pymavlink import mavutil
import time


timeout = 5

# Establish the connection
connection = mavutil.mavlink_connection('COM5', baud=57600)

try:
    # Wait
    connection.wait_heartbeat(blocking=True, timeout=timeout)
    print("Heartbeat received from system")

    # Send a text
    text_message = "Radar starts, "  # Ensure this is a string

    # Send the message as byte
    connection.mav.statustext_send(mavutil.mavlink.MAV_SEVERITY_INFO, text_message.encode('utf-8'))

    print("Text message sent:", text_message)

except KeyboardInterrupt:
    print("Operation interrupted by user")

except Exception as e:
    print(f"Error: {e}")
```
